```python
import pandas as pd
import numpy as np

path = "/Users/ianlent/Downloads/Historical Surprise WMT.O (1).xlsx"

raw = pd.read_excel(
    path,
    header=None,
    skiprows=10  # you said there are 6 comment rows
)

raw.head(10)
```

```python
raw = raw.rename(columns={0: "metric"})
raw.head(10)
```

```python
import pandas as pd
import numpy as np

# raw already loaded with skiprows=6 and renamed col 0 -> "metric"
# raw = raw.rename(columns={0: "metric"})

# 1) Grab the header row that contains "Surprise" / "Report Period" / Q1 Apr-20 ...
hdr = raw.loc[raw["metric"].astype(str).str.strip().eq("Surprise")].iloc[0]

# 2) Period labels start at column index 2 (since col 1 is 'Report Period')
period_labels = hdr.iloc[2:].astype(str).str.strip()

# 3) Keep only columns whose period label is not NaN/blank/"nan"
valid_cols = period_labels[
    period_labels.notna() &
    (period_labels != "") &
    (period_labels.str.lower() != "nan")
].index

# mapping: numeric col index -> period string like "Q1 Apr-20"
col_map = {c: period_labels.loc[c] for c in valid_cols}

# 4) Pull the metric rows you care about
metrics_needed = ["Surprise %", "Actual", "Mean", "7d Price Reaction"]

df_wide = raw[raw["metric"].isin(metrics_needed)].copy()

# Keep only metric + the valid period columns, then rename to period labels
df_wide = df_wide[["metric"] + list(valid_cols)].rename(columns=col_map)

# 5) Transpose into event rows
events = df_wide.set_index("metric").T.reset_index().rename(columns={"index": "report_period"})

# 6) Clean column names and convert to numeric
events = events.rename(columns={
    "Surprise %": "surprise_pct",
    "Actual": "actual_eps",
    "Mean": "mean_estimate",
    "7d Price Reaction": "price_reaction_7d",
})

for c in ["surprise_pct", "actual_eps", "mean_estimate", "price_reaction_7d"]:
    events[c] = pd.to_numeric(events[c], errors="coerce")

# Optional: parse a real month from the period label (Apr-20 etc.)
events["period_month"] = (
    events["report_period"].astype(str)
    .str.extract(r"([A-Za-z]{3}-\d{2})", expand=False)
)
events["period_month"] = pd.to_datetime(events["period_month"], format="%b-%y", errors="coerce")

events = events.sort_values("period_month").reset_index(drop=True)

events.head(8), events.tail(8)
```

```python
cutoff = pd.Timestamp.today() - pd.DateOffset(years=5)

wmt5 = events[events["period_month"] >= cutoff].copy()

wmt5[["report_period","period_month","surprise_pct","price_reaction_7d"]].head(), wmt5.shape
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.scatter(wmt5["surprise_pct"], wmt5["price_reaction_7d"])
plt.axhline(0)
plt.axvline(0)
plt.title("WMT: EPS Surprise vs 7D Price Reaction")
plt.xlabel("EPS Surprise (%)")
plt.ylabel("7D Price Reaction (%)")
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import numpy as np

ticker = "WMT"
px_raw = yf.download(ticker, start="2019-01-01", auto_adjust=True, progress=False)

# If yfinance returned multiindex columns, flatten them
if isinstance(px_raw.columns, pd.MultiIndex):
    px_raw.columns = [c[0] for c in px_raw.columns]  # keep first level like 'Close'

px = px_raw.reset_index().rename(columns={"Date":"date", "Close":"close"})
px = px[["date","close"]].sort_values("date").reset_index(drop=True)

px.head()
```

```python
earn = events.copy()
# Extract "Apr-20" from strings like "Q1 Apr-20"
earn["month_token"] = (
    earn["report_period"].astype(str)
    .str.extract(r"([A-Za-z]{3}-\d{2})", expand=False)
)

earn["period_month"] = pd.to_datetime(earn["month_token"], format="%b-%y", errors="coerce")

earn[["report_period","month_token","period_month"]].head(10)
```

```python
earn["event_date"] = earn["period_month"] + pd.offsets.Day(15)

earn["event_date"] = pd.to_datetime(earn["event_date"]).astype("datetime64[ns]")
px["date"] = pd.to_datetime(px["date"]).astype("datetime64[ns]")

earn = earn.sort_values("event_date").reset_index(drop=True)
px = px.sort_values("date").reset_index(drop=True)

earn_px = pd.merge_asof(
    earn,
    px,
    left_on="event_date",
    right_on="date",
    direction="backward"
)

earn_px[["report_period", "event_date", "close","surprise_pct","price_reaction_7d"]].head(10)
```

```python
earn = earn.copy()

# Keep only quarterly rows
earn_q = earn[earn["report_period"].astype(str).str.startswith("Q")].copy()

# If you have period_month, great; if not, rebuild it
if "period_month" not in earn_q.columns or earn_q["period_month"].isna().all():
    earn_q["month_token"] = (
        earn_q["report_period"].astype(str)
        .str.extract(r"([A-Za-z]{3}-\d{2})", expand=False)
    )
    earn_q["period_month"] = pd.to_datetime(earn_q["month_token"], format="%b-%y", errors="coerce")

earn_q = earn_q.sort_values("period_month").reset_index(drop=True)

# Labels
earn_q["miss"] = earn_q["surprise_pct"] < 0
earn_q["beat"] = earn_q["surprise_pct"] > 0

earn_q[["report_period","period_month","surprise_pct","price_reaction_7d"]].head()
```

```python
plt.figure(figsize=(14,5))
plt.plot(earn_q["period_month"], earn_q["price_reaction_7d"], marker="o", label="7D Price Reaction")
plt.axhline(0)
plt.title("WMT: 7-Day Price Reaction After Earnings (Refinitiv)")
plt.ylabel("7D reaction")
plt.grid(True)
plt.legend()
plt.show()

# Color bars by beat/miss
colors = np.where(earn_q["surprise_pct"] >= 0, "#2ca02c", "#d62728")  # green / red

plt.figure(figsize=(14,5))

plt.bar(
    earn_q["period_month"],
    earn_q["surprise_pct"],
    width=45,            # width in DAYS → much more visible
    color=colors,
    alpha=0.85,
    edgecolor="black",
    linewidth=0.4
)

plt.axhline(0, color="black", linewidth=0.8)
plt.title("WMT: EPS Surprise (%) by Quarter")
plt.ylabel("EPS Surprise (%)")
plt.xlabel("Report Period")
plt.grid(axis="y", alpha=0.3)

plt.show()
```

```python
import statsmodels.api as sm

df_reg = earn_q.dropna(subset=["surprise_pct","price_reaction_7d"]).copy()
X = sm.add_constant(df_reg["surprise_pct"])
y = df_reg["price_reaction_7d"]

model = sm.OLS(y, X).fit()
model.summary()
```
